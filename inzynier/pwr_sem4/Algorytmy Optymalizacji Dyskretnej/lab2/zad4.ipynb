```python
# autor: Łukasz Machnik
from pulp import LpMaximize, LpMinimize, LpProblem, LpStatus, lpSum, LpVariable
```

```python
m = 8
n = 8
k = 2
numberOfConteners = 8
contenerLocation = [[0 for j in range(n)] for i in range(m)]
contenerLocation[0][5] = 1
contenerLocation[0][6] = 1
contenerLocation[2][4] = 1
contenerLocation[4][2] = 1
contenerLocation[4][6] = 1
contenerLocation[4][7] = 1
contenerLocation[5][0] = 1
contenerLocation[6][4] = 1
for r in contenerLocation:
    print(r)
```

```python
model = LpProblem("rozmieszczenie_kamer", sense=LpMinimize)
cameraLocation = [[LpVariable("kamera(%s, %s)" % (i, j), cat="Binary") for j in range(n)] for i in range(m)]
for i in range(m):
    for j in range(n):
        if contenerLocation[i][j] == 1:
            model += (cameraLocation[i][j] == 0, "occupied(%s, %s)" % (i, j))
            camerasInRange = 0
            northBound = max(i - k, 0)
            eastBound = min(j + k, n - 1)
            southBound = min(i + k, m - 1)
            westBound = max(j - k, 0)
            for s in range(northBound, i):
                camerasInRange += cameraLocation[s][j]
            for s in range(j, eastBound):
                camerasInRange += cameraLocation[i][s]
            for s in range(i, southBound + 1):
                camerasInRange += cameraLocation[s][j]
            for s in range(westBound, j):
                camerasInRange += cameraLocation[i][s]
            model += (camerasInRange >= 1, "require_camera_for(%s, %s)" % (i, j))
model += lpSum(cameraLocation)
model.solve()
print(model)
```

```python
print("Liczba wymaganych kamer: %s" % model.objective.value())
for i in range(m):
    line = ""
    for j in range(n):
        if(contenerLocation[i][j] == 1):
            line += "X"
        elif(cameraLocation[i][j].value() == 1):
            line += "C"
        else:
            line += "-"
    print(line)
```
